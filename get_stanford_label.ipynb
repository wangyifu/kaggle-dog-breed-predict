```python
from os import walk

f = []

result = []

for (_, dirnames, _) in walk('input/Images'):
    for dirname in dirnames:
        breed_name = dirname[dirname.index('-')+1:].lower()
        for (_, _, filenames) in walk('input/Images/' + dirname):
            for filename in filenames:
                result.append({'id': filename.split('.')[0], 'breed': breed_name})

print(result)
```

```python
import csv

with open('input/Images/stanford_labels.csv', 'w') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=['id','breed'])
    writer.writeheader()
    for row_dict in result:
        writer.writerow(row_dict)
```

```python
breed_set = set()
for (_, dirnames, _) in walk('input/Images'):
    for dirname in dirnames:
        breed_name = dirname[dirname.index('-')+1:].lower()
        breed_set.add(breed_name)
print(breed_set)
```
